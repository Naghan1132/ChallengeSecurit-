```python
import pandas as pd
import numpy as np


pd.set_option('display.max_columns', None)

# Chemin vers votre fichier .log
chemin_fichier = "C:/Users/nagrimault/Downloads/log_brute.log"

# Nombre de lignes à sauter
#nombre_lignes_a_sauter = 11950738 #TEST
nombre_lignes_a_sauter = 7377860

# Créer une liste de dictionnaires pour stocker les données
donnees = []


# Ouvrir le fichier en mode lecture
with open(chemin_fichier, 'r') as f:
    # Lire et ignorer les premières lignes jusqu'à la ligne spécifiée
    for _ in range(nombre_lignes_a_sauter):
        next(f)
    
    # Lire et traiter les lignes restantes
    for ligne in f:
        # Diviser la ligne en fonction de votre structure de données
        elements = ligne.split()

    
        # Créer un dictionnaire avec les informations de la ligne
        ligne_donnees = {
            'Mois': elements[0] if len(elements) > 0 else '',
            'Jour': elements[1] if len(elements) > 1 else '',
            'Heure': elements[2] if len(elements) > 2 else '',
            'Ip': elements[3] if len(elements) > 3 else '',
            'A enlever': elements[4] if len(elements) > 4 else '', # kernel à enlever
            'Kernel': elements[5] if len(elements) > 5 else '', 
            'Status': elements[6] if len(elements) > 6 else '',
            'FW': elements[7] if len(elements) > 7 else '',
            'Rule': elements[8] if len(elements) > 8 else '',
            'In': elements[9] if len(elements) > 9 else '',
            'Out': elements[10] if len(elements) > 10 else '',
            'Mac': elements[11] if len(elements) > 11 else '',
            'SRC': elements[12] if len(elements) > 12 else '',
            'Dist': elements[13] if len(elements) > 13 else '',
            'Len': elements[14] if len(elements) > 14 else '',
            'Tos': elements[15] if len(elements) > 15 else '',
            'Prec': elements[16] if len(elements) > 16 else '',
            'TTL': elements[17] if len(elements) > 17 else '',
            'ID': elements[18] if len(elements) > 18 else '',
            'DF': elements[19] if 'DF' in elements else 'None',
            'Proto': elements[20] if 'DF' in elements else elements[19],
            'SPT': elements[21] if 'DF' in elements else elements[20],
            'DPT': elements[22] if 'DF' in elements else elements[21]
            # 'Window': elements[23] if elements[6]=="PERMIT" else '',
            # 'Res': elements[24] if elements[6]=="PERMIT" else '',
            # 'ColSyn': elements[25] if elements[6]=="PERMIT" else '',
            # 'Urgp': elements[26] if elements[6]=="PERMIT" else '',
        }
        
        donnees.append(ligne_donnees)

# Créer un DataFrame à partir de la liste de dictionnaires
df = pd.DataFrame(donnees)

# Appliquer la fonction lambda pour extraire la partie après "=" uniquement pour la colonne "MAC"
#df['Mac'] = df['Mac'].apply(lambda x: x.split('=')[1] if isinstance(x, str) else x)
#df['FW'] = df['FW'].apply(lambda x: x.split('=')[1] if isinstance(x, str) else x)
df['Rule'] = df['Rule'].apply(lambda x: x.split('=')[1] if isinstance(x, str) else x)
df['In'] = df['In'].apply(lambda x: x.split('=')[1] if isinstance(x, str) else x)
df['Out'] = df['Out'].apply(lambda x: x.split('=')[1] if isinstance(x, str) else x)
df['SRC'] = df['SRC'].apply(lambda x: x.split('=')[1] if isinstance(x, str) else x)
df['Dist'] = df['Dist'].apply(lambda x: x.split('=')[1] if isinstance(x, str) else x)
#df['Tos'] = df['Tos'].apply(lambda x: x.split('=')[1] if isinstance(x, str) else x)
#df['Prec'] = df['Prec'].apply(lambda x: x.split('=')[1] if isinstance(x, str) else x)
#df['TTL'] = df['TTL'].apply(lambda x: x.split('=')[1] if isinstance(x, str) else x)
#df['ID'] = df['ID'].apply(lambda x: x.split('=')[1] if isinstance(x, str) else x)

df['SPT'] = df['SPT'].apply(lambda x: x.split('=')[1] if isinstance(x, str) else x)
df['DPT'] = df['DPT'].apply(lambda x: x.split('=')[1] if isinstance(x, str) else x)

#df['Window'] = df['Window'].apply(lambda x: x.split('=')[1] if isinstance(x, str) else x)
#df['Res'] = df['Res'].apply(lambda x: x.split('=')[1] if isinstance(x, str) else x)
#df['Len'] = df['Len'].str.split('=').str[1]
df['Proto'] = df['Proto'].str.split('=').str[1] 
#df['Urgp'] = df['Urgp'].str.split('=').str[1]


col_a_drop = ["A enlever","Kernel","Ip","Mac","FW","Len","TTL","ID","Tos","Prec","DF"]

df.drop(columns=col_a_drop, inplace=True)
```
